```python
# définissez l'ID de l'application
app_id = 'YOUR_APP_ID'

# définissez l'URL de l'API
base_url = 'https://api.deezer.com/search?q=YEAR:2022&output=json&app_id=' + app_id

# créez une liste pour stocker les données
data_list = []

# boucle sur chaque jour de l'année
for day in range(1, 366):
    # définissez le paramètre de jour
    day_param = '&day=' + str(day)
    
    # effectuez la requête
    response = requests.get(base_url + day_param)
    
    # récupérez les données
    data_list.append(response.json())

# convertir les données en DataFrame


# afficher le DataFrame
```

```python

df = pd.DataFrame(data_list[0]['data'])
for elt in data_list[1:-1]:
    df= pd.concat([df,pd.DataFrame(elt['data'])])
```
