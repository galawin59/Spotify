# isolation forest

```python
X= df_elias.drop(columns=['popularity','genre'],axis=1)
y= df_elias[['popularity']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True, test_size=0.15, random_state=42 )
```

```python
from sklearn.ensemble import IsolationForest 

# Instanciation de l'algorithme 
clf = IsolationForest(n_estimators = 100, max_samples='auto', contamination=float(0.01), max_features=1.0, bootstrap=False, n_jobs=-1, random_state=42, verbose=0) 

# Entrainement du modèle 
clf.fit(X) 

# Prédictions 
y_pred = clf.predict(X)
```

```python
df_pred = pd.DataFrame(y_pred)
```

```python
df_pred.value_counts()
```

```python
df_elias['isolation']= y_pred
```

```python
df_elias = df_elias[df_elias['isolation']==1]
```

# argument de setup
1. data: Le jeu de données à utiliser pour l'exécution des fonctionnalités de pycaret.

2. target: Le nom de la colonne de la cible à prédire (y).

3. session_id: Le nom d'un identifiant de session à utiliser pour créer et enregistrer des modèles.

4. log_experiment: Booléen qui indique si le journal des expériences doit être enregistré.

5. experiment_name: Le nom de l'expérience à enregistrer.

6. log_plots: Booléen qui indique si les graphiques doivent être enregistrés.

7. log_profile: Booléen pour activer le profilage des données.

8. profile: Profilage des données à activer.

9. normalize: Booléen qui indique si le jeu de données doit être normalisé.

10. transformation: Transformation des données à appliquer.

11. ignore_low_variance: Booléen qui indique si les variables ayant une faible variabilité doivent être ignorées.

12. ignore_high_correlation: Booléen qui indique si les variables hautement corrélées doivent être ignorées.

13. feature_selection: Sélection des caractéristiques à appliquer.

14. pca: Dimensionnalité réduite à appliquer.

15. pca_method: Méthode de l'ACP à appliquer.

16. fix_imbalance: Booléen qui indique si l'imbalance des classes doit être corrigée.

17. sample_frac: Fréquence d'échantillonnage pour la correction d'imbalance des classes.

18. polynomial_features: Booléen qui indique si des caractéristiques polynomiales doivent être générées.

19. polynomial_degree: Degré du polynôme à générer.
